#A docstring (documentation string) is a string written inside a function, class, or module to explain what it does.

It helps humans understand the code, not the computer.

In Python, docstring are written using triple quotes (""" """) or ('''  ''')

#WHY docstring are used 

     1.Explain purpose of the code
     2.Describe parameters and return values
     3.Make code readable & professional
     4.Used by tools like help() and documentation generators
