# Variable 

:- Learn to store value in container for later use. Variables is a concept in programming that allows us to give a label to a piece of data so that we can refer or reference that data using the chosen variable name. We will see in this lesson how to create variable and to use the variable to access the contained value.


#Why do we use variables?

:-We use variables to store values and give them meaningful names.

Example:

    Suppose 7321830841 is a phone number. If we see only the number, we do not know whose number it is.
    
    But when we write:

```python
nisha = 7321830841 
print(nisha)
```

-Now we understand that this is Nisha’s phone number.

In the same way, in programming languages, variables are used to store different types of data such as numbers, strings, and other values, so that the program becomes easy to understand and use.
