# PrettyTbale

* What is PrettyTable?
  
  :- PrettyTable is a Python library used to display   data in a clean table format in the terminal (console).

* Why Do We Use PrettyTable?

  :-We use PrettyTable when:

    1. We want to show data in rows and columns
    2. We want clean formatting in terminal
    3. We are working on CLI projects
    4. We want better readability

```python
# Code without PrettyTable

name = ["Nisha Thakur", "Shivam Thakur"]
age  = [21, 22]

print("name age")
for i in range(2):
    print(name[i], age[1]) # output :- This looks simple and not properly aligned.
```

```python
# Code with PrettyTable

from prettytable import PrettyTable

table = PrettyTable()

table.field_name = ["Name", "Age"] # it's define columns 

table.add.row[("Nisha", 21)]
table.add.row[("Shivam", 22)]
```

```python
# code with PrettyTble 
# columns 

from prettytable import PrettyTable

table1 = PrettyTable()
table1.field_names = ["name","age","section","salary"]
print(table1)
```

```python
# Define Row's and coloumns both 

from prettytable import PrettyTable

table2 = PrettyTable() #intial the table 
table2.field_names = ["Name of student","E.number","section", "course"]
table2.add_row(["Nisha Thakur","A3046","A","BCA"])
table2.add_row(["Shivam Thakur","A3043","A","BSC"])
print(table2)
```

```python
# Add columns method 

table3 = PrettyTable()

column_name = ["name", "age", "state"]
table3.add_column(column_name[0], ["nisha thakur", "Shivam Thakur", "Aaditya Thakur"])
table3.add_column(column_name[1],  ["21", "22", "17"])
table3.add_column(column_name[2], ["delhi", "delhi", "Up"])
print(table3)
```

```python
# get_String() method to print the table 
x = table3.get_string(fields =["name", "state"])
print(x)
```

```python
# select specific row in the table 

y= table3.get_string(start=1, end =2)
print(y)
                    
```

```python
# Sort the table increassing order 

table3.sortby = "age"
print(table3)
```

```python
# sort the table by decreassing order 

table3.sortby = "age"
table3.reversesort = True
print(table3)
```

```python
# delete row from the table 
table2.del_row(0)
print(table2)
```

```python
# Trunk The Table if you want :- if you trunk the table all the row are delete but still coluns are exist.

table2.clear_rows()
print(table2)
```
