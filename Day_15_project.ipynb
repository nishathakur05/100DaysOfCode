```python
#code without constructors but with functions
class faculty:
    def putdata(self): # method1 
        self.id= int(input("enter faculty id"))
        self.name=input("enter name")
        self.salary =float(input("enter faculty salary"))
    def display(self): #method2
        print("faculty is:",self.id)
        print("faculty name:",self.name)
        print("faculty salary",self.salary)

a=faculty()
a.putdata()
a.display()
```

```python
class faculty1:
    def __init__(self): # method1 
        self.id= int(input("enter faculty id"))
        self.name=input("enter name")
        self.salary =float(input("enter faculty salary"))
    def display(self): #method2
        print("faculty is:",self.id)
        print("faculty name:",self.name)
        print("faculty salary",self.salary)

a=faculty1()
# a.putdata()
a.display()
```

```python
%%writefile coffee_maker.py

class CoffeeMaker:
    
    def __init__(self):
        self.resources = {
            "water": 300,
            "milk": 200,
            "coffee": 100
        }

    def report(self):
        print(f"Water: {self.resources['water']}ml")
        print(f"Milk: {self.resources['milk']}ml")
        print(f"Coffee: {self.resources['coffee']}g")
```

```python
#step2:
from coffee_maker import CoffeeMaker

coffee_maker = CoffeeMaker()
coffee_maker.report()
```
