# Default argmanets 

:-jab hum function define karte time kisi parameter ko pehle se ek value de dete hain, to function call karte waqt agar hum us argument ko pass na karein, to default value use ho jati hai.

:-A default argument is a value that a function uses automatically if you don’t give that value while calling the function.

```python
def greet(name="User"):
    print("Hello",name)

greet("Nisha")  #output : Hello Nisha
greet()         #output:  Hello user
```

# Note:-
  
    name="User" → "User" is the default argument
    Agar hum greet() me kuch pass nahi karte, to "User" use hota hai
    Agar hum value pass karte hain, to wahi value use hoti hai
